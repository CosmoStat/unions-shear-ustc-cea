```python
from astropy.io import fits
from astropy import units
import matplotlib.pylab as plt
```

```python
idx = 0
n_split = 1
weight = 'w'

input_files = [
    f'LF/ggl_agn_{idx}_n_split_{n_split}_{weight}_angular.fits',
    f'LF/ggl_agn_{idx}_n_split_{n_split}_{weight}_physical.fits',
]

colors = [
    'r',
    'blue',
]    
```

```python
dat = []
for inpf in input_files:
    this_dat = fits.getdata(inpf)
    dat.append(this_dat)
```

```python
d_ang_mean = 1507.40 * units.Mpc
```

```python
plt.loglog(
    dat[0]['meanr'],
    dat[0]['gamT'],
    'o',
    color=colors[0],
    label='angular stack',
)
#plt.loglog(dat[0]['r_nom'], dat[0]['gamT'], 'x', color=colors[0], label='')

x = dat[1]['meanr']
plt.loglog(
    x,
    dat[1]['gamT'],
    'd-',
    label='physical stack',
)

x = ((dat[1]['r_nom'] * units.Mpc / d_ang_mean) * units.rad).to('arcmin')
#plt.loglog(
#    x,
#    dat[1]['gamT'],
#    's',
#    label=r'physical stack, $r=\bar\theta / \bar d_<$',
#)

plt.legend()
plt.xlabel(r'$\theta$ [arcmin]')
_ = plt.ylabel(r'$\gamma_{\rm t}(\theta)$')

plt.savefig('g_t_ang_phys_test.png')
```

```python
key = 'npairs'

plt.loglog(
    dat[0]['meanr'],
    dat[0][key],
    'o',
    color=colors[0],
    label='angular stack',
)

x = dat[1]['meanr']
plt.loglog(
    x,
    dat[1][key],
    'd',
    label='physical stack',
)

plt.legend()
plt.xlabel(r'$\theta$ [arcmin]')
plt.ylabel(r'$n_{\rm pair}(\theta)$')

_ = plt.savefig(f'{key}_ang_phys_test.png')
```

```python
key = 'weight'

plt.loglog(
    dat[0]['meanr'],
    dat[0][key],
    'o',
    color=colors[0],
    label='angular stack',
)

x = dat[1]['meanr']
plt.loglog(
    x,
    dat[1][key],
    'd',
    label='physical stack',
)

plt.legend()
plt.xlabel(r'$\theta$ [arcmin]')
plt.ylabel(r'$w(\theta)$')

plt.savefig(f'{key}_ang_phys_test.png')
```

```python
r = ((dat[0]['r_nom'] * units.arcmin).to('rad') * d_ang_mean).value
print(r[0], r[-1])
plt.loglog(r, dat[0]['gamT'], 'o')
plt.loglog(r, -dat[0]['gamT'], 'o', markersize=5)

r = ((dat[0]['meanr'] * units.arcmin).to('rad') * d_ang_mean).value
print(r[0], r[-1])
plt.loglog(r, dat[0]['gamT'], 'd')
plt.loglog(r, -dat[0]['gamT'], 'd', markersize=5)

plt.xlabel(r'$r$ [Mpc]')
_ = plt.ylabel(r'$\gamma_{\rm t}(r)$')

r = dat[1]['r_nom']
print(r[0], r[-1])
plt.loglog(r, dat[1]['gamT'], 'x')
plt.loglog(r, -dat[1]['gamT'], 'x', markersize=5)

r = dat[1]['meanr']
print(r[0], r[-1])
plt.loglog(r, dat[1]['gamT'], 's')
_ = plt.loglog(r, -dat[1]['gamT'], 's', markersize=5)
```
